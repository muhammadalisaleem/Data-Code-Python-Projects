```
# Ali Saleem _ 23L-2638 _ BDS-3B
# Intro to Data Science Assignment # 1 p2

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DF_Iris = pd.read_csv('Iris.csv')

#Analyzing the SepalLengthCm Attribute

print("Sepal Length column:\n", DF_Iris['SepalLengthCm'])

print("\nData type of SepalLengthCm:", DF_Iris['SepalLengthCm'].dtype)

print("Total number of instances in SepalLengthCm:", DF_Iris['SepalLengthCm'].count())

print("Number of null values in SepalLengthCm:", DF_Iris['SepalLengthCm'].isnull().sum())

print("Minimum value in SepalLengthCm:", DF_Iris['SepalLengthCm'].min())
print("Maximum value in SepalLengthCm:", DF_Iris['SepalLengthCm'].max())

print("Mode of SepalLengthCm:", DF_Iris['SepalLengthCm'].mode()[0])
print("Mean of SepalLengthCm:", DF_Iris['SepalLengthCm'].mean())
print("Median of SepalLengthCm:", DF_Iris['SepalLengthCm'].median())

print("Variance of SepalLengthCm:", DF_Iris['SepalLengthCm'].var())
print("Standard deviation of SepalLengthCm:", DF_Iris['SepalLengthCm'].std())

# Identifying outliers using the IQR method
Q1 = DF_Iris['SepalLengthCm'].quantile(0.25)
Q3 = DF_Iris['SepalLengthCm'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = DF_Iris[(DF_Iris['SepalLengthCm'] < lower_bound) | (DF_Iris['SepalLengthCm'] > upper_bound)]
print("\nOutliers in SepalLengthCm:\n", outliers['SepalLengthCm'])
print("Number of outliers in SepalLengthCm:", outliers.shape[0])


# Histogram for SepalLengthCm, SepalWidthCm, and PetalLengthCm
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(DF_Iris['SepalLengthCm'], bins=20, color='lightblue', edgecolor='black')
plt.title('Sepal Length')

plt.subplot(1, 3, 2)
plt.hist(DF_Iris['SepalWidthCm'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Sepal Width')

plt.subplot(1, 3, 3)
plt.hist(DF_Iris['PetalLengthCm'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Petal Length')

plt.tight_layout()
plt.show()

# Box plot for SepalLengthCm, SepalWidthCm, and PetalLengthCm
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.boxplot(y=DF_Iris['SepalLengthCm'], color='lightblue')
plt.title('Box Plot - Sepal Length')

plt.subplot(1, 3, 2)
sns.boxplot(y=DF_Iris['SepalWidthCm'], color='lightgreen')
plt.title('Box Plot - Sepal Width')

plt.subplot(1, 3, 3)
sns.boxplot(y=DF_Iris['PetalLengthCm'], color='lightcoral')
plt.title('Box Plot - Petal Length')

plt.tight_layout()
plt.show()

# Scatter plot for SepalLengthCm vs SepalWidthCm
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=DF_Iris, palette='deep')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()


# Print unique species labels and their counts
species_labels = DF_Iris['Species'].unique()
species_counts = DF_Iris['Species'].value_counts()

print("\nUnique Species Labels:", species_labels)
print("\nSpecies Counts:\n", species_counts)
```
