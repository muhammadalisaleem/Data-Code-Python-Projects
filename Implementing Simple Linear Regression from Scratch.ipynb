```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading the dataset
dataset = '/content/avocado.csv'
df = pd.read_csv(dataset)

x = df['Total Volume'].values
y = df['AveragePrice'].values

def calculate_slope_intercept(x, y):
    
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x)**2)


    slope = numerator / denominator
    intercept = mean_y - slope * mean_x

    return slope, intercept

def predict(x, slope, intercept):
    return slope * x + intercept

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def plot_regression_line(x, y, slope, intercept):

    y_pred = predict(x, slope, intercept)

    plt.scatter(x, y, color="blue", label="Actual data points")

    plt.plot(x, y_pred, color="red", label="Regression line")

    plt.xlabel('Total Volume')
    plt.ylabel('Average Price')
    plt.legend()
    plt.show()


slope, intercept = calculate_slope_intercept(x, y)

y_pred = predict(x, slope, intercept)

mse = mean_squared_error(y, y_pred)

print(f"Mean Squared Error: {mse}")

plot_regression_line(x, y, slope, intercept)
```
